```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np 
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Bimodal 3D distribution 
mean1 = [-2, -1]
cov1 = [[0.1, 0], [0, 0.5]]
data1 = np.random.multivariate_normal(mean1, cov1, 500)

# Second peak centered at (2, 2)
mean2 = [0, 0]
cov2 = [[0.5, 0], [0, 0.5]]
data2 = np.random.multivariate_normal(mean2, cov2, 500)

# Combine into a single dataset
data = np.vstack([data1, data2])
x, y = data[:, 0], data[:, 1]
```

```python
plt.figure()
plt.scatter(x,y)
plt.show() 
```

```python
# Multivariate kernel here 
def guassian(x, xi, H):
    d = x - xi
    inv_H = np.linalg.inv(H)
    det_H = np.linalg.det(H)
    
    norm_const = 1 / (2 * np.pi * np.sqrt(det_H))
    exponent = -0.5 * np.dot(d.T, np.dot(inv_H, d))    
    return norm_const * np.exp(exponent)

H = 0.3**2 * np.cov(data.T)

print(guassian(x,x,H))
```
